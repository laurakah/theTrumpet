```python
import time

from selenium import webdriver
from selenium.webdriver.common.keys import Keys

browser = webdriver.Chrome()

browser.get("https://factba.se/transcripts/speeches")
time.sleep(1)

elem = browser.find_element_by_tag_name("body")

no_of_pagedowns = 1000

while no_of_pagedowns:
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
    no_of_pagedowns-=1

a_tags = browser.find_elements_by_css_selector('h3 a')

links = []

for a in a_tags:
     links.append(a.get_attribute('href'))
```

```python
speech_parts = []

for url in links:
    res = requests.get(url)
    res_bytes = res.content
    res_str = res.text

    soup_out_of_text = BeautifulSoup(res_str, "lxml")
    soup_out_of_bytes = BeautifulSoup(res_bytes, "html.parser")

    soup = soup_out_of_bytes
    
    speech_part = soup.find_all("div", "topic-media-row")
    speech_parts.append(speech_part)
```

```python
text = ""

for parts in speech_parts:
    for part in parts:
        speaker = part.find("div", class_="speaker-label")
        if speaker.text == "Donald Trump":
            text_part = part.find("a").text
            text = text + text_part
```

```python
text_file = open("../../data/trump_speeches.txt", "w")
n = text_file.write(text)
text_file.close()
```
